```python
# imports
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag as pt
from nltk.stem import PorterStemmer as PS
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer as WNL
import re, sys, string
sys.setrecursionlimit(0x100000)
```

```python
# reads filename
def read_operation(file_nm):
    if file_nm is None:
        print("Error!.. No file found!")
        exit()
    with open(file_nm) as handle:
        mydata = handle.read()
    return mydata
```

```python
# tokenize word
def word_tokenized(to_token):
    if to_token is None:
        print("Error!.. No handle specified for tokenizing")
    generated_t = word_tokenize(to_token)
    return generated_t
```

```python
# Question 1. (a)

f_one = 'process.txt'
f_one_data = read_operation(f_one)
f_one_tokened = word_tokenized(f_one_data)
print(f_one_tokened)
```

```python
# Question 1. (b)
'''
Normalisation consists of :-
#lowercase
#stop word removal
#white spaces removal
#abbreviations
#sparse terms, particular words

>>> to pending <<<
report this answer as your answer
'''
def normalisation(raw_d):
    to_normalise = f_one_data
    to_normalise_lower = to_normalise.lower()
    to_normalise_r = to_normalise_lower.replace(",", "").replace("'", "").replace(";", "").replace(".","").replace('"', "")
    to_word_tokenised = word_tokenize(to_normalise_r)
    #remove stop words
    sw = set(stopwords.words('english'))
    filtered_text = [i for i in to_word_tokenised if not i in sw]
    print(filtered_text)
    return filtered_text
normalised_final = normalisation(f_one_data)
```

```python
def mapper(nltk_tag):
    if nltk_tag.startswith('J'): return 'a'
    elif nltk_tag.startswith('V'): return 'v'
    elif nltk_tag.startswith('N'): return 'n'    
    elif nltk_tag.startswith('R'): return 'r'
    else: return 'n'
```
